```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import models
```

```
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
# load the dataset
path = "/content/drive/My Drive/Colab Notebooks/trial_train.csv"
dataset=pd.read_csv(path)
# split into input (X) and output (y) variables
train_features = dataset.copy()
train_labels = train_features.pop('DAMAGE_AMOUNT_ORDER_TOTAL')
model = keras.Sequential([
      layers.Dense(60,input_dim=3, activation='relu'),
      layers.Dense(30, activation='relu'),
      layers.Dense(1)
  ])
model.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.Adam(0.001), metrics=['accuracy'])
model.fit(train_features, train_labels, epochs=150, batch_size=5)
model.save_weights("model.tf")
```
